### Kernel Density Estimation



```python
import numpy
from matplotlib import pyplot
import chaospy

numpy.random.seed(42)
```

```python
samples = chaospy.Uniform(0, 1).sample((2, 7))
distribution = chaospy.GaussianKDE(samples)
distribution
```

```python
mesh = numpy.mgrid[0:1:50j, 0:1:50j]
pyplot.contourf(mesh[0], mesh[1], distribution.pdf(mesh))
pyplot.colorbar()
pyplot.scatter(*samples)
pyplot.show()
```

It is also possible to manually define the bandwidth:

```python
distribution = chaospy.GaussianKDE(samples, bandwidth2=0.1**2)
pyplot.contourf(mesh[0], mesh[1], distribution.pdf(mesh))
pyplot.colorbar()
pyplot.scatter(*samples)
pyplot.show()
```

Anisotropic bandwidths for each dimension:

```python
distribution = chaospy.GaussianKDE(samples, bandwidth2=[0.1**2, 0.3**2])
pyplot.contourf(mesh[0], mesh[1], distribution.pdf(mesh))
pyplot.colorbar()
pyplot.scatter(*samples)
pyplot.show()
```

Covariance structure is also possible:

```python
bandwidth2 = [[0.1**2, -0.008], [-0.008, 0.1**2]]
distribution = chaospy.GaussianKDE(samples, bandwidth2)
pyplot.contourf(mesh[0], mesh[1], distribution.pdf(mesh))
pyplot.colorbar()
pyplot.scatter(*samples)
pyplot.show()
```
